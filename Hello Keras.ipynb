# Hello Keras!
[Keras](https://keras.io)是一个可以将Tensorflow或Theano作为后端的深度学习框架。由于它的高度模块化，写神经网络会方便很多。

下面，我将用Keras实现一个简单的异或神经网络，作为Keras的HelloWorld。

## 引入库

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers.core import Activation, Dense
```

## 训练

```python
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[0],[1],[1],[0]], "float32")

model = Sequential()
model.add(Dense(32, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])

model.fit(training_data, target_data, epochs=1000, verbose=2)

print model.predict(training_data)
```

```python
def generate():
    res = [ ]
    x = np.linspace(0, 1)
    for xx in x:
        y = np.linspace(0, 1)
        for yy in y:
            res.append([xx, yy])
    return res
```

```python
%matplotlib inline
data = generate()
res = model.predict(data)
falsex = [ ]
falsey = [ ]
truex = [ ]
truey = [ ]
edgex = [ ]
edgey = [ ]
for i in range(0, len(res)):
    if res[i][0] > 0.5:
        truex.append(data[i][0])
        truey.append(data[i][1])
    if res[i][0] < 0.5:
        falsex.append(data[i][0])
        falsey.append(data[i][1])
    if res[i][0] > 0.49 and res[i][0] < 0.51:
        edgex.append(data[i][0])
        edgey.append(data[i][1])
plt.plot(falsex, falsey, '.')
plt.plot(truex, truey, '.')
plt.plot(edgex, edgey, 'o')
```
