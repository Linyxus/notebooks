### 欧几里得算法(辗转相除法)
算法基于这样的事实：
$$ (a, b) = (b, a \ mod \ b) $$
且又有
$$ (a, 0) = a $$

```python
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)
```

```python
%timeit gcd(np.random.randint(10000000), np.random.randint(10000000))
```

gcd存在如下性质：
$$ (a, b) = (a, b + ax) $$
其中 $ x \in Z $。
换而言之，令$f(b)=gcd(a,b)，则 f(b)以a为周期。$

```python
%matplotlib inline
x = np.arange(100)
g = gcd(16, 24)
gs = np.full(100, g)
y = [ ]
for xx in x:
    y.append(gcd(16, 24 + xx))
plt.plot(x, gs)
plt.plot(x, y)
```
